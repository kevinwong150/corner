```python
import csv

with open('1920_team_id.csv', 'w+', newline='') as csvfile:
    csv_writer = csv.writer(csvfile, delimiter=',', quoting=csv.QUOTE_MINIMAL)
    csv_writer.writerow(['team_id'] + ['full_name'] + [ 'name' ] + [ 'tournament_id' ] + [ 'tournament_name' ])
    for team_id in sorted(teams_dict.keys(), key=lambda x: int(x)):
        csv_writer.writerow(
            [ team_id ] + 
            [ teams_dict[team_id]['full_name'] ] + 
            [ teams_dict[team_id]['name'] ] + 
            [ teams_dict[team_id]['tournament_id'] ] +
            [ teams_dict[team_id]['tournament_name'] ]
        )
```
